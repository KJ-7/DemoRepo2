```python
# Write a code to analyse expenses over time
Let us say your expense for every month are listed below,
January - 2200
February - 2350
March - 2600
April - 2130
May - 2190
Create a list to store these monthly expenses and using that find out,

1. In Feb, how many dollars you spent extra compare to January?
2. Find out your total expense in first quarter (first three months) of the year.
3. Find out if you spent exactly 2000 dollars in any month
4. June month just finished and your expense is 1980 dollar. Add this item to our monthly expense list
5. You returned an item that you bought in a month of April and
got a refund of 200$. Make a correction to your monthly expense list
based on this
```

```python
expenses = [
    ["January", 2200],
    ["February", 2350],
    ["March", 2600],
    ["April", 2130],
    ["May", 2190],
    ["July", 2000]
]

```

```python
# 1 In Feb, how many dollars you spent extra compare to January?
extra_dollars = expenses[1][1]-expenses[0][1]
extra_dollars
```

```python
# 2 Find out your total expense in first quarter (first three months) of the year.
total_expenses = expenses[0][1]+expenses[1][1]+expenses[2][1]
total_expenses
```

```python
# 3 Find out if you spent exactly 2000 dollars in any month
any2000 = False
month2000 = ""
for n in range(len(expenses)):
    if (expenses[n][1] == 2000):
        any2000 = True
        month2000 = expenses[n][0]
        break
if any2000:
  print("There is a month you spent exactly 2000, and it was in "+ month2000)
else:
  print("There is no month you spent exactly 2000")
```

```python
# 4 June month just finished and your expense is 1980 dollar. Add this item to our monthly expense list
expenses.append(["June", 1980])
expenses
```

```python
# 5. You returned an item that you bought in a month of April and
# got a refund of 200$. Make a correction to your monthly expense list
# based on this
for n in range(len(expenses)):
    if expenses[n][0] == "April":
        expenses[n][1] += 200
expenses
```
